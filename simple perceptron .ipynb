```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

by sklearn.databases is a set of databases
load_iris is a dataset of iris value into variable "iris"

```python
iris
```

```python
x=iris.data[:,(2,3)]
```

iris data slice in 2 and 3 in to variable "x"

```python
x
```

```python
iris.target
```

our perceptron target is to get things....0,1,2

```python
y=(iris.target==0)
```

```python
y
```

tern in to true and false statement 

```python
y=(iris.target==0).astype(np.int)
```

```python
y
```

```python
x
```

```python
from sklearn.linear_model import Perceptron
```

importing Perceptron from sklearn.linear_modle

```python
per_clf=Perceptron (random_state=42)
```

```python
per_clf.fit(x,y)
```

```python
y_pred=per_clf.predict(x)
```

predict the values

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

accuracy_score tells us how accurate our prediction...it has 2 values(1.0=all correct,0=all wrong)

```python
accuracy_score(y,y_pred)
```
